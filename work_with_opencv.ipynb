```python
from google.colab.patches import cv2_imshow
import cv2
img=cv2.imread('./1.jpg')
cv2_imshow(img)
cv2.waitKey()
cv2.destroyAllWindows()
```

همچنین میتوانیم از متپلات لیب برای نمایش تصاویر استفاده کرد

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

# جای کانال رنگ ها عوض شده و باید آن را درست کنیم
# از روش نامپای اسلایسینگ استفاده کردیم

```python
plt.imshow(img[:,:,::-1])
```

این کار را هم میتوان انجام داد

```python
plt.imshow(img[...,::-1])
```

# یا میتوان از خود اوپن سی وی استفاده کرد

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

متغییر هایی که با حروف بزرگ نوشته میشود یک عدد ثابت هستند

```python
print(cv2.COLOR_BGR2RGB)
```
